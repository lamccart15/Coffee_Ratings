```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read csv and load df
df = pd.read_csv("data/coffee_data_cleaned.csv")
df
```

```python
#Basic Data Cleaning
df = df.drop(columns=['Unnamed: 0'])
df
```

## Full Model

```python
# Set features. This will also be used as your x values.
selected_features = df[['aroma', 'flavor', 'aftertaste', 'acidity', 'body', 'balance', 'uniformity', 'clean_cup', 
                        'sweetness', 'moisture', 'category_one_defects', 'category_two_defects']]
```

```python
#Set y and Split into Test/Train
from sklearn.model_selection import train_test_split

y = df["cupper_points"]

X_train, X_test, y_train, y_test = train_test_split(
    selected_features, y, random_state=1)
```

```python
X_train.head()
```

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Create Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model = model.fit(X_train_scaled, y_train)
```

```python
#Train Model
training_score = model.score(X_train_scaled, y_train)
testing_score = model.score(X_test_scaled, y_test) 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Plot the Residuals for the Training and Testing data

plt.scatter(model.predict(X_train_scaled), model.predict(X_train_scaled) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test_scaled), model.predict(X_test_scaled) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
plt.savefig('assets/img/residual_plot.png')
```

## Reduced Model

```python
# Set features. This will also be used as your x values.
selected_features2 = df[['flavor', 'aftertaste', 'balance']]
```

```python
#Set y and Split into Test/Train
from sklearn.model_selection import train_test_split

y2 = df["cupper_points"]

X_train2, X_test2, y_train2, y_test2 = train_test_split(
    selected_features2, y2, random_state=1)
```

```python
X_train2.head()
```

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train2)
X_train_scaled2 = X_scaler.transform(X_train2)
X_test_scaled2 = X_scaler.transform(X_test2)
```

```python
#Create Model
from sklearn.linear_model import LinearRegression
model_reduced = LinearRegression()
model_reduced = model_reduced.fit(X_train_scaled2, y_train2)
```

```python
#Train Model
training_score_reduced = model_reduced.score(X_train_scaled2, y_train2)
testing_score_reduced = model_reduced.score(X_test_scaled2, y_test2) 

print(f"Training Score: {training_score_reduced}")
print(f"Testing Score: {testing_score_reduced}")
```

```python
# Plot the Residuals for the Training and Testing data

plt.scatter(model_reduced.predict(X_train_scaled2), model_reduced.predict(X_train_scaled2) - y_train2, c="blue", label="Training Data")
plt.scatter(model_reduced.predict(X_test_scaled2), model_reduced.predict(X_test_scaled2) - y_test2, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y2.min(), xmax=y2.max())
plt.title("Residual Plot")
plt.savefig('assets/img/residual_plot_reduced.png')
```
