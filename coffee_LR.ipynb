```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read csv and load df
df = pd.read_csv("data/coffee_data_cleaned.csv")
df
```

```python
#Basic Data Cleaning
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
# Set features. This will also be used as your x values.
selected_features = df[['aroma', 'flavor', 'aftertaste', 'acidity', 'body', 'balance', 'uniformity', 'clean_cup', 'sweetness', 'moisture']]
```

```python
#Set y and Split into Test/Train
from sklearn.model_selection import train_test_split

y = df["cupper_points"]

X_train, X_test, y_train, y_test = train_test_split(
    selected_features, y, random_state=1)
```

```python
X_train.head()
```

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Create Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=2000)
classifier
```
